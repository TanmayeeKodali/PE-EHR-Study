```python
#Loading the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# For .xlsx files (Excel)
df = pd.read_excel('data/pe_dataset.xlsx', sheet_name='Dataset')

print(df.shape)  # Shows (number_of_rows, number_of_columns)
print(df.head())  # Shows first 5 rows
```

```python
# 2.1: Collapsed age groups for analysis
def categorize_age(Age_groups):
    """Collapse 9 age categories into 4 research categories"""
    if Age_groups in [1, 2, 3]:
        return '<40'
    elif Age_groups in [4, 5]:
        return '40-60'
    elif Age_groups == 6:
        return '60-75'
    elif Age_groups in [7, 8, 9]:
        return '>75'
    else:
        return np.nan

df['Age_Category'] = df['Age_groups'].apply(categorize_age)
print(f"✓ Created Age_Category variable")

# 2.2: Create count of additional diagnostic procedures (excluding CTPA)
# Count non-CTPA procedures: CT_chest + VP_scan + PulmonaryAngio
df['Additional_Procedures'] = (
    df['CT_chest'].fillna(0) + 
    df['VP_scan'].fillna(0) + 
    df['PulmonaryAngio'].fillna(0)
)

# 2.3: Create binary "Excessive Diagnostics" variable (>1 additional test)
df['Excessive_Diagnostics'] = (df['Additional_Procedures'] > 1).astype(int)
print(f"✓ Created Additional_Procedures and Excessive_Diagnostics variables")

# 2.4: Create consolidated race/ethnicity variable (collapse small groups if needed)
# This is a placeholder - you may want to collapse categories based on sample sizes
df['Race_Ethnicity'] = 'Unknown'  # Initialize

# Map race/ethnicity (modify based on your actual column names)
df.loc[df['Non-Hispanic White'] == 1, 'Race_Ethnicity'] = 'Non-Hispanic White'
df.loc[df['Non-Hispanic Black'] == 1, 'Race_Ethnicity'] = 'Non-Hispanic Black'
df.loc[df['Non-Hispanic Asian'] == 1, 'Race_Ethnicity'] = 'Non-Hispanic Asian'
df.loc[df['Hispanic/Latinx'] == 1, 'Race_Ethnicity'] = 'Hispanic/Latinx'
df.loc[df['American Indian or Alaska Native'] == 1, 'Race_Ethnicity'] = 'American Indian/Alaska Native'
df.loc[df['Native Hawaiian or Other Pacific Islander'] == 1, 'Race_Ethnicity'] = 'Native Hawaiian/Pacific Islander'
df.loc[(df['Non-Hispanic Other or Unknown'] == 1) | 
       (df['Other races/ethnicities'] == 1) | 
       (df['Unknown/Declined race/ethnicity'] == 1), 'Race_Ethnicity'] = 'Other/Unknown'

print(f"✓ Created Race_Ethnicity variable")
```

```python
# AGE GROUP AND PE INCIDENCE

print("\n" + "="*80)
print("RESEARCH QUESTION 1: PE INCIDENCE BY AGE GROUP")
print("="*80)

# Create crosstab of age groups and PE status
age_pe_crosstab = pd.crosstab(
    df['Age_Category'], 
    df['PE_in_index_hospitalization '],
    margins=True
)
age_pe_crosstab.columns = ['PE Negative', 'PE Positive', 'Total']
print("\n--- Age Group × PE Status Crosstabulation ---")
print(age_pe_crosstab)

# Calculate PE incidence rates by age group
print("\n--- PE Incidence Rates by Age Group ---")
age_pe_rates = df.groupby('Age_Category')['PE_in_index_hospitalization '].agg([
    ('N', 'count'),
    ('PE_Positive', 'sum'),
    ('PE_Rate_%', lambda x: x.mean() * 100)
]).round(2)
print(age_pe_rates)
```

```python
# HYPERTENSION AND PE (OVERALL & STRATIFIED BY AGE)

print("\n" + "="*80)
print("RESEARCH QUESTION 2: HYPERTENSION AND PE")
print("="*80)

# 5.1: Overall hypertension prevalence by PE status
print("\n--- Overall: Hypertension × PE Status ---")
htn_pe_crosstab = pd.crosstab(
    df['Hypertension'],
    df['PE_in_index_hospitalization '],
    margins=True
)
htn_pe_crosstab.columns = ['PE Negative', 'PE Positive', 'Total']
htn_pe_crosstab.index = ['No Hypertension', 'Hypertension', 'Total']
print(htn_pe_crosstab)

# Calculate percentages
print("\n--- Hypertension Prevalence by PE Status (%) ---")
htn_prev_by_pe = df.groupby('PE_in_index_hospitalization ')['Hypertension'].agg([
    ('N', 'count'),
    ('HTN_N', 'sum'),
    ('HTN_Prevalence_%', lambda x: x.mean() * 100)
]).round(2)
htn_prev_by_pe.index = ['PE Negative', 'PE Positive']
print(htn_prev_by_pe)

# 5.2: Hypertension stratified by age groups
print("\n--- Hypertension × PE Status STRATIFIED BY AGE GROUP ---")
for age_cat in ['<40', '40-60', '60-75', '>75']:
    age_subset = df[df['Age_Category'] == age_cat]
    print(f"\n  Age Group: {age_cat}")
    print(f"  Total N: {len(age_subset)}")
    
    htn_pe_age = pd.crosstab(
        age_subset['Hypertension'],
        age_subset['PE_in_index_hospitalization ']
    )
    htn_pe_age.columns = ['PE Negative', 'PE Positive']
    htn_pe_age.index = ['No HTN', 'HTN']
    print(htn_pe_age)
    
    # PE rates by HTN status within this age group
    pe_rate_no_htn = age_subset[age_subset['Hypertension']==0]['PE_in_index_hospitalization '].mean() * 100
    pe_rate_with_htn = age_subset[age_subset['Hypertension']==1]['PE_in_index_hospitalization '].mean() * 100
    print(f"  PE rate (No HTN): {pe_rate_no_htn:.1f}%")
    print(f"  PE rate (HTN): {pe_rate_with_htn:.1f}%")
```

```python
#DIAGNOSTIC INTENSITY BY RACE (PE+ PATIENTS ONLY)

print("\n" + "="*80)
print("RESEARCH QUESTION 3: DIAGNOSTIC INTENSITY BY RACE (PE+ ONLY)")
print("="*80)

# Filter to PE+ patients only
df_pe_positive = df[df['PE_in_index_hospitalization '] == 1].copy()
print(f"Analyzing {len(df_pe_positive)} PE+ patients\n")

# 6.1: Distribution of diagnostic procedures in PE+ patients
print("\n--- Distribution of Additional Diagnostic Procedures (PE+ patients) ---")
print(df_pe_positive['Additional_Procedures'].value_counts().sort_index())

print("\n--- Excessive Diagnostics (>1 additional test) ---")
excessive_n = df_pe_positive['Excessive_Diagnostics'].sum()
excessive_pct = df_pe_positive['Excessive_Diagnostics'].mean() * 100
print(f"  N with excessive diagnostics: {excessive_n} ({excessive_pct:.1f}%)")

# 6.2: Diagnostic procedures by race/ethnicity
print("\n--- Additional Procedures by Race/Ethnicity (PE+ patients) ---")
race_procedures = df_pe_positive.groupby('Race_Ethnicity')['Additional_Procedures'].agg([
    ('N', 'count'),
    ('Mean', 'mean'),
    ('SD', 'std'),
    ('Median', 'median'),
    ('Min', 'min'),
    ('Max', 'max')
]).round(2)
print(race_procedures)

# 6.3: Excessive diagnostics by race/ethnicity
print("\n--- Excessive Diagnostics by Race/Ethnicity (PE+ patients) ---")
race_excessive = df_pe_positive.groupby('Race_Ethnicity')['Excessive_Diagnostics'].agg([
    ('N', 'count'),
    ('N_Excessive', 'sum'),
    ('Percent_Excessive', lambda x: x.mean() * 100)
]).round(2)
print(race_excessive)

# 6.4: Individual procedure types by race/ethnicity
print("\n--- Individual Procedure Types by Race/Ethnicity (PE+ patients) ---")
procedures = ['CTPA', 'CT_chest', 'VP_scan', 'PulmonaryAngio']
for proc in procedures:
    print(f"\n  {proc}:")
    proc_by_race = df_pe_positive.groupby('Race_Ethnicity')[proc].agg([
        ('N', 'count'),
        ('N_Yes', 'sum'),
        ('Percent', lambda x: x.mean() * 100)
    ]).round(2)
    print(proc_by_race)
```

```python
#OTHER DESCRIPTIVE STATISTICS (For Table 1)

print("\n" + "="*80)
print("ADDITIONAL DESCRIPTIVE STATISTICS")
print("="*80)

# 7.1: Sex distribution
print("\n--- Sex Distribution ---")
print(df['Sex'].value_counts())
sex_by_pe = pd.crosstab(df['Sex'], df['PE_in_index_hospitalization '], normalize='columns') * 100
sex_by_pe.columns = ['PE Negative %', 'PE Positive %']
print("\nSex Distribution by PE Status (%):")
print(sex_by_pe.round(1))

# 7.2: Other comorbidities (descriptive only)
print("\n--- Other Comorbidities (Descriptive) ---")
comorbidities = ['Diabetes', 'History of CVA', 'CAD', 'Heart Failure', 
                 'Dialysis', 'Prior Hx of VTE', 'COVID-19 in Prior 30-days', 'PAD']

for comorb in comorbidities:
    if comorb in df.columns:
        print(f"\n  {comorb}:")
        overall_prev = df[comorb].mean() * 100
        pe_pos_prev = df[df['PE_in_index_hospitalization ']==1][comorb].mean() * 100
        pe_neg_prev = df[df['PE_in_index_hospitalization ']==0][comorb].mean() * 100
        print(f"    Overall: {overall_prev:.1f}%")
        print(f"    PE Positive: {pe_pos_prev:.1f}%")
        print(f"    PE Negative: {pe_neg_prev:.1f}%")

# 7.3: Race/ethnicity distribution
print("\n--- Race/Ethnicity Distribution ---")
print(df['Race_Ethnicity'].value_counts())
race_by_pe = pd.crosstab(df['Race_Ethnicity'], df['PE_in_index_hospitalization '])
race_by_pe.columns = ['PE Negative', 'PE Positive']
print("\nRace/Ethnicity by PE Status:")
print(race_by_pe)
```

```python
#CREATE SUMMARY TABLE (Table 1 for manuscript)

print("\n" + "="*80)
print("TABLE 1: BASELINE CHARACTERISTICS BY PE STATUS")
print("="*80)

def create_summary_table(df):
    """Create comprehensive Table 1"""
    
    pe_neg = df[df['PE_in_index_hospitalization '] == 0]
    pe_pos = df[df['PE_in_index_hospitalization '] == 1]
    
    print(f"\n{'Variable':<40} {'Overall':<15} {'PE Negative':<15} {'PE Positive':<15}")
    print(f"{'':<40} {'N={len(df)}':<15} {'N={len(pe_neg)}':<15} {'N={len(pe_pos)}':<15}")
    print("-" * 85)
    
    # Age groups
    print("\nAge Category, n (%)")
    for age in ['<40', '40-60', '60-75', '>75']:
        overall = (df['Age_Category'] == age).sum()
        overall_pct = (df['Age_Category'] == age).mean() * 100
        neg = (pe_neg['Age_Category'] == age).sum()
        neg_pct = (pe_neg['Age_Category'] == age).mean() * 100
        pos = (pe_pos['Age_Category'] == age).sum()
        pos_pct = (pe_pos['Age_Category'] == age).mean() * 100
        print(f"  {age:<38} {overall} ({overall_pct:.1f}%){'':<5} {neg} ({neg_pct:.1f}%){'':<5} {pos} ({pos_pct:.1f}%)")
    
    # Sex
    print("\nSex, n (%)")
    for sex in df['Sex'].unique():
        overall = (df['Sex'] == sex).sum()
        overall_pct = (df['Sex'] == sex).mean() * 100
        neg = (pe_neg['Sex'] == sex).sum()
        neg_pct = (pe_neg['Sex'] == sex).mean() * 100
        pos = (pe_pos['Sex'] == sex).sum()
        pos_pct = (pe_pos['Sex'] == sex).mean() * 100
        print(f"  {sex:<38} {overall} ({overall_pct:.1f}%){'':<5} {neg} ({neg_pct:.1f}%){'':<5} {pos} ({pos_pct:.1f}%)")
    
    # Hypertension (KEY VARIABLE)
    print("\nHypertension, n (%)")
    overall = df['Hypertension'].sum()
    overall_pct = df['Hypertension'].mean() * 100
    neg = pe_neg['Hypertension'].sum()
    neg_pct = pe_neg['Hypertension'].mean() * 100
    pos = pe_pos['Hypertension'].sum()
    pos_pct = pe_pos['Hypertension'].mean() * 100
    print(f"  Yes{'':<35} {overall} ({overall_pct:.1f}%){'':<5} {neg} ({neg_pct:.1f}%){'':<5} {pos} ({pos_pct:.1f}%)")
    
    # Other comorbidities
    print("\nOther Comorbidities, n (%)")
    for comorb in ['Diabetes', 'History of CVA', 'CAD']:
        if comorb in df.columns:
            overall = df[comorb].sum()
            overall_pct = df[comorb].mean() * 100
            neg = pe_neg[comorb].sum()
            neg_pct = pe_neg[comorb].mean() * 100
            pos = pe_pos[comorb].sum()
            pos_pct = pe_pos[comorb].mean() * 100
            print(f"  {comorb:<38} {overall} ({overall_pct:.1f}%){'':<5} {neg} ({neg_pct:.1f}%){'':<5} {pos} ({pos_pct:.1f}%)")

create_summary_table(df)
```

```python
# COMPREHENSIVE SUMMARY STATISTICS TABLE - ALL VARIABLES

print("\n" + "="*80)
print("COMPREHENSIVE SUMMARY STATISTICS: ALL VARIABLES")
print("="*80)

def create_comprehensive_summary():
    """Create comprehensive summary statistics for ALL variables in dataset"""
    
    pe_neg = df[df['PE_in_index_hospitalization '] == 0]
    pe_pos = df[df['PE_in_index_hospitalization '] == 1]
    
    print(f"\n{'Variable':<45} {'Overall':<20} {'PE Negative':<20} {'PE Positive':<20}")
    print(f"{'':<45} {'N={len(df)}':<20} {'N={len(pe_neg)}':<20} {'N={len(pe_pos)}':<20}")
    print("=" * 105)
    
    # DEMOGRAPHICS
    print("\n" + "="*105)
    print("DEMOGRAPHICS")
    print("=" * 105)
    
    # Age groups (detailed - all 9 categories)
    print("\nAge Groups (Original Categories), n (%)")
    for age_val in sorted(df['Age_groups'].dropna().unique()):
        overall = (df['Age_groups'] == age_val).sum()
        overall_pct = (df['Age_groups'] == age_val).mean() * 100
        neg = (pe_neg['Age_groups'] == age_val).sum()
        neg_pct = (pe_neg['Age_groups'] == age_val).mean() * 100
        pos = (pe_pos['Age_groups'] == age_val).sum()
        pos_pct = (pe_pos['Age_groups'] == age_val).mean() * 100
        
        if age_val == 1: age_label = "  18-25 years"
        elif age_val == 2: age_label = "  25-35 years"
        elif age_val == 3: age_label = "  35-45 years"
        elif age_val == 4: age_label = "  45-55 years"
        elif age_val == 5: age_label = "  55-65 years"
        elif age_val == 6: age_label = "  65-75 years"
        elif age_val == 7: age_label = "  75-85 years"
        elif age_val == 8: age_label = "  85-95 years"
        elif age_val == 9: age_label = "  95+ years"
        else: age_label = f"  Group {age_val}"
        
        print(f"{age_label:<45} {overall} ({overall_pct:.1f}%){'':<8} {neg} ({neg_pct:.1f}%){'':<8} {pos} ({pos_pct:.1f}%)")
    
    # Age categories (collapsed)
    print("\nAge Categories (Collapsed), n (%)")
    for age in ['<40', '40-60', '60-75', '>75']:
        overall = (df['Age_Category'] == age).sum()
        overall_pct = (df['Age_Category'] == age).mean() * 100
        neg = (pe_neg['Age_Category'] == age).sum()
        neg_pct = (pe_neg['Age_Category'] == age).mean() * 100
        pos = (pe_pos['Age_Category'] == age).sum()
        pos_pct = (pe_pos['Age_Category'] == age).mean() * 100
        print(f"  {age:<43} {overall} ({overall_pct:.1f}%){'':<8} {neg} ({neg_pct:.1f}%){'':<8} {pos} ({pos_pct:.1f}%)")
    
    # Sex
    print("\nSex, n (%)")
    for sex in ['Female', 'Male']:
        if sex in df['Sex'].values:
            overall = (df['Sex'] == sex).sum()
            overall_pct = (df['Sex'] == sex).mean() * 100
            neg = (pe_neg['Sex'] == sex).sum()
            neg_pct = (pe_neg['Sex'] == sex).mean() * 100
            pos = (pe_pos['Sex'] == sex).sum()
            pos_pct = (pe_pos['Sex'] == sex).mean() * 100
            print(f"  {sex:<43} {overall} ({overall_pct:.1f}%){'':<8} {neg} ({neg_pct:.1f}%){'':<8} {pos} ({pos_pct:.1f}%)")
    
    # Race/Ethnicity
    print("\nRace/Ethnicity, n (%)")
    race_vars = ['Non-Hispanic White', 'Non-Hispanic Black', 'Non-Hispanic Asian', 
                 'Hispanic/Latinx', 'American Indian or Alaska Native', 
                 'Native Hawaiian or Other Pacific Islander', 'Non-Hispanic Other or Unknown',
                 'Other races/ethnicities', 'Unknown/Declined race/ethnicity']
    
    for race_var in race_vars:
        if race_var in df.columns:
            overall = df[race_var].sum()
            overall_pct = df[race_var].mean() * 100
            neg = pe_neg[race_var].sum()
            neg_pct = pe_neg[race_var].mean() * 100
            pos = pe_pos[race_var].sum()
            pos_pct = pe_pos[race_var].mean() * 100
            print(f"  {race_var:<43} {overall} ({overall_pct:.1f}%){'':<8} {neg} ({neg_pct:.1f}%){'':<8} {pos} ({pos_pct:.1f}%)")
    
    # COMORBIDITIES
    print("\n" + "="*105)
    print("COMORBIDITIES")
    print("=" * 105)
    
    comorbidities = {
        'Diabetes': 'Diabetes',
        'Hypertension': 'Hypertension',
        'History of CVA': 'History of CVA',
        'CAD': 'Coronary Artery Disease',
        'Heart Failure': 'Heart Failure',
        'Dialysis': 'Dialysis',
        'Prior Hx of VTE': 'Prior VTE History',
        'COVID-19 in Prior 30-days': 'COVID-19 (Prior 30 days)',
        'PAD': 'Peripheral Artery Disease'
    }
    
    print("\nComorbidities Present, n (%)")
    for var, label in comorbidities.items():
        if var in df.columns:
            overall = df[var].sum()
            overall_pct = df[var].mean() * 100
            neg = pe_neg[var].sum()
            neg_pct = pe_neg[var].mean() * 100
            pos = pe_pos[var].sum()
            pos_pct = pe_pos[var].mean() * 100
            print(f"  {label:<43} {overall} ({overall_pct:.1f}%){'':<8} {neg} ({neg_pct:.1f}%){'':<8} {pos} ({pos_pct:.1f}%)")
    
    # PE DIAGNOSIS CHARACTERISTICS
    print("\n" + "="*105)
    print("PE DIAGNOSIS CHARACTERISTICS")
    print("=" * 105)
    
    print("\nDischarge Diagnosis Group, n (%)")
    for group in sorted(df['Group'].dropna().unique()):
        overall = (df['Group'] == group).sum()
        overall_pct = (df['Group'] == group).mean() * 100
        neg = (pe_neg['Group'] == group).sum()
        neg_pct = (pe_neg['Group'] == group).mean() * 100 if len(pe_neg) > 0 else 0
        pos = (pe_pos['Group'] == group).sum()
        pos_pct = (pe_pos['Group'] == group).mean() * 100 if len(pe_pos) > 0 else 0
        
        if group == 'Group 1': group_label = "  Group 1 (Principal Dx)"
        elif group == 'Group 2': group_label = "  Group 2 (Secondary Dx)"
        elif group == 'Group 3': group_label = "  Group 3 (No PE codes)"
        else: group_label = f"  {group}"
        
        print(f"{group_label:<45} {overall} ({overall_pct:.1f}%){'':<8} {neg} ({neg_pct:.1f}%){'':<8} {pos} ({pos_pct:.1f}%)")
    
    # PE subtypes (among PE+ patients only)
    print("\nPE Subtypes (Among PE+ Patients Only), n (%)")
    if 'IfPE_OnlySubsegmental' in df.columns:
        subseg_n = pe_pos['IfPE_OnlySubsegmental'].sum()
        subseg_pct = pe_pos['IfPE_OnlySubsegmental'].mean() * 100 if len(pe_pos) > 0 else 0
        print(f"  {'Subsegmental PE only':<43} N/A{'':<16} N/A{'':<16} {subseg_n} ({subseg_pct:.1f}%)")
    
    if 'IfPE_CorPulmonale' in df.columns:
        cor_pulm_n = pe_pos['IfPE_CorPulmonale'].sum()
        cor_pulm_pct = pe_pos['IfPE_CorPulmonale'].mean() * 100 if len(pe_pos) > 0 else 0
        print(f"  {'Cor pulmonale present':<43} N/A{'':<16} N/A{'':<16} {cor_pulm_n} ({cor_pulm_pct:.1f}%)")
    
    # DIAGNOSTIC PROCEDURES
    print("\n" + "="*105)
    print("DIAGNOSTIC PROCEDURES PERFORMED")
    print("=" * 105)
    
    procedures = {
        'CTPA': 'CT Pulmonary Angiography (CTPA)',
        'CT_chest': 'Chest CT',
        'VP_scan': 'Ventilation-Perfusion Scan',
        'PulmonaryAngio': 'Pulmonary Angiography',
        'RadiologyConsult': 'Radiology Consultation'
    }
    
    print("\nProcedures Performed, n (%)")
    for var, label in procedures.items():
        if var in df.columns:
            overall = df[var].sum()
            overall_pct = df[var].mean() * 100
            neg = pe_neg[var].sum()
            neg_pct = pe_neg[var].mean() * 100
            pos = pe_pos[var].sum()
            pos_pct = pe_pos[var].mean() * 100
            print(f"  {label:<43} {overall} ({overall_pct:.1f}%){'':<8} {neg} ({neg_pct:.1f}%){'':<8} {pos} ({pos_pct:.1f}%)")
    
    # Additional procedures count (derived variable)
    print("\nAdditional Procedures Count (excluding CTPA)")
    print(f"  {'Mean (SD)':<43}", end='')
    overall_mean = df['Additional_Procedures'].mean()
    overall_sd = df['Additional_Procedures'].std()
    neg_mean = pe_neg['Additional_Procedures'].mean()
    neg_sd = pe_neg['Additional_Procedures'].std()
    pos_mean = pe_pos['Additional_Procedures'].mean()
    pos_sd = pe_pos['Additional_Procedures'].std()
    print(f"{overall_mean:.2f} ({overall_sd:.2f}){'':<8} {neg_mean:.2f} ({neg_sd:.2f}){'':<8} {pos_mean:.2f} ({pos_sd:.2f})")
    
    print(f"  {'Median [IQR]':<43}", end='')
    overall_med = df['Additional_Procedures'].median()
    overall_q1 = df['Additional_Procedures'].quantile(0.25)
    overall_q3 = df['Additional_Procedures'].quantile(0.75)
    neg_med = pe_neg['Additional_Procedures'].median()
    neg_q1 = pe_neg['Additional_Procedures'].quantile(0.25)
    neg_q3 = pe_neg['Additional_Procedures'].quantile(0.75)
    pos_med = pe_pos['Additional_Procedures'].median()
    pos_q1 = pe_pos['Additional_Procedures'].quantile(0.25)
    pos_q3 = pe_pos['Additional_Procedures'].quantile(0.75)
    print(f"{overall_med:.1f} [{overall_q1:.1f}-{overall_q3:.1f}]{'':<5} {neg_med:.1f} [{neg_q1:.1f}-{neg_q3:.1f}]{'':<5} {pos_med:.1f} [{pos_q1:.1f}-{pos_q3:.1f}]")
    
    print("\nExcessive Diagnostics (>1 additional procedure), n (%)")
    overall = df['Excessive_Diagnostics'].sum()
    overall_pct = df['Excessive_Diagnostics'].mean() * 100
    neg = pe_neg['Excessive_Diagnostics'].sum()
    neg_pct = pe_neg['Excessive_Diagnostics'].mean() * 100
    pos = pe_pos['Excessive_Diagnostics'].sum()
    pos_pct = pe_pos['Excessive_Diagnostics'].mean() * 100
    print(f"  {'Yes':<43} {overall} ({overall_pct:.1f}%){'':<8} {neg} ({neg_pct:.1f}%){'':<8} {pos} ({pos_pct:.1f}%)")
    
    # COR PULMONALE DETECTION METHODS (among those with cor pulmonale)
    if 'IfPE_CorPulmonale' in df.columns:
        df_cor_pulm = pe_pos[pe_pos['IfPE_CorPulmonale'] == 1].copy()
        if len(df_cor_pulm) > 0:
            print("\n" + "="*105)
            print(f"COR PULMONALE DETECTION METHODS (Among {len(df_cor_pulm)} PE+ patients with cor pulmonale)")
            print("=" * 105)
            
            cor_pulm_methods = {
                'IfCorPulmonale_CT': 'Detected on CT',
                'IfCorPulmonale_Echo': 'Detected on Echocardiography',
                'IfCorPulmonale_Troponin': 'Troponin elevation'
            }
            
            for var, label in cor_pulm_methods.items():
                if var in df.columns:
                    n = df_cor_pulm[var].sum()
                    pct = df_cor_pulm[var].mean() * 100
                    print(f"  {label:<43} {n} ({pct:.1f}%)")
    
    print("\n" + "="*105)

# Call the comprehensive summary function
create_comprehensive_summary()
```

```python
#FIGURE 1: GROUPED BAR CHART - PE INCIDENCE BY AGE AND HYPERTENSION (Research Questions 1 & 2)

def calculate_proportion_ci(successes, total, confidence=0.95):
    """Calculate Wilson score confidence interval for proportions"""
    if total == 0:
        return 0, 0, 0
    
    p = successes / total
    z = stats.norm.ppf((1 + confidence) / 2)
    
    denominator = 1 + z**2 / total
    centre = (p + z**2 / (2 * total)) / denominator
    margin = z * np.sqrt((p * (1 - p) + z**2 / (4 * total)) / total) / denominator
    
    lower = max(0, centre - margin)
    upper = min(1, centre + margin)
    
    return p, lower, upper

# Prepare data for plotting
age_order = ['<40', '40-60', '60-75', '>75']
plot_data = []

for age in age_order:
    age_subset = df[df['Age_Category'] == age]
    
    # No hypertension group
    no_htn = age_subset[age_subset['Hypertension'] == 0]
    n_no_htn = len(no_htn)
    pe_no_htn = no_htn['PE_in_index_hospitalization '].sum()
    rate_no_htn, ci_lower_no_htn, ci_upper_no_htn = calculate_proportion_ci(pe_no_htn, n_no_htn)
    
    plot_data.append({
        'Age': age,
        'Hypertension': 'No',
        'PE_Rate': rate_no_htn * 100,
        'CI_Lower': ci_lower_no_htn * 100,
        'CI_Upper': ci_upper_no_htn * 100,
        'N': n_no_htn
    })
    
    # Hypertension group
    yes_htn = age_subset[age_subset['Hypertension'] == 1]
    n_yes_htn = len(yes_htn)
    pe_yes_htn = yes_htn['PE_in_index_hospitalization '].sum()
    rate_yes_htn, ci_lower_yes_htn, ci_upper_yes_htn = calculate_proportion_ci(pe_yes_htn, n_yes_htn)
    
    plot_data.append({
        'Age': age,
        'Hypertension': 'Yes',
        'PE_Rate': rate_yes_htn * 100,
        'CI_Lower': ci_lower_yes_htn * 100,
        'CI_Upper': ci_upper_yes_htn * 100,
        'N': n_yes_htn
    })

plot_df = pd.DataFrame(plot_data)

# Create the figure
fig, ax = plt.subplots(figsize=(12, 7))

# Define positions and width
x_pos = np.arange(len(age_order))
width = 0.35

# Separate data for each hypertension group
no_htn_data = plot_df[plot_df['Hypertension'] == 'No']
yes_htn_data = plot_df[plot_df['Hypertension'] == 'Yes']

# Create bars
bars1 = ax.bar(x_pos - width/2, no_htn_data['PE_Rate'], width, 
               label='No Hypertension', color='#3498db', alpha=0.8, edgecolor='black', linewidth=0.5)

bars2 = ax.bar(x_pos + width/2, yes_htn_data['PE_Rate'], width, 
               label='Hypertension', color='#e74c3c', alpha=0.8, edgecolor='black', linewidth=0.5)

# Add error bars (95% CI)
for i, (idx, row) in enumerate(no_htn_data.iterrows()):
    error_lower = row['PE_Rate'] - row['CI_Lower']
    error_upper = row['CI_Upper'] - row['PE_Rate']
    ax.errorbar(x_pos[i] - width/2, row['PE_Rate'], 
                yerr=[[error_lower], [error_upper]], 
                fmt='none', ecolor='black', capsize=4, capthick=1.5)

for i, (idx, row) in enumerate(yes_htn_data.iterrows()):
    error_lower = row['PE_Rate'] - row['CI_Lower']
    error_upper = row['CI_Upper'] - row['PE_Rate']
    ax.errorbar(x_pos[i] + width/2, row['PE_Rate'], 
                yerr=[[error_lower], [error_upper]], 
                fmt='none', ecolor='black', capsize=4, capthick=1.5)

# Add sample size annotations inside bars
for i, (idx, row) in enumerate(no_htn_data.iterrows()):
    ax.text(x_pos[i] - width/2, row['PE_Rate']/2, f"n={int(row['N'])}", 
            ha='center', va='center', fontsize=8, fontweight='bold', color='white')

for i, (idx, row) in enumerate(yes_htn_data.iterrows()):
    ax.text(x_pos[i] + width/2, row['PE_Rate']/2, f"n={int(row['N'])}", 
            ha='center', va='center', fontsize=8, fontweight='bold', color='white')

# Customize plot
ax.set_xlabel('Age Group (years)', fontsize=12, fontweight='bold')
ax.set_ylabel('PE Incidence Rate (%)', fontsize=12, fontweight='bold')
ax.set_title('Pulmonary Embolism Incidence by Age Group and Hypertension Status', 
             fontsize=14, fontweight='bold', pad=20)
ax.set_xticks(x_pos)
ax.set_xticklabels(age_order, fontsize=11)
ax.legend(loc='upper left', fontsize=11, frameon=True, shadow=True)
ax.set_ylim(0, max(plot_df['CI_Upper']) * 1.15)  # Add 15% space above highest CI
ax.grid(axis='y', alpha=0.3, linestyle='--')

# Add subtle background
ax.set_facecolor('#f8f9fa')

plt.tight_layout()
plt.savefig('Figure1_PE_by_Age_HTN.png', dpi=300, bbox_inches='tight')
print("✓ Figure 1 saved as 'Figure1_PE_by_Age_HTN.png'")
plt.show()

# Print summary statistics for Figure 1
print("\nFigure 1 Data Summary:")
print(plot_df.to_string(index=False))
```

```python
# FIGURE 2: BOX PLOT WITH SWARM OVERLAY - DIAGNOSTIC INTENSITY BY RACE (Research Question 3)

# Filter to PE+ patients only
df_pe_pos = df[df['PE_in_index_hospitalization '] == 1].copy()

# Remove groups with very small sample sizes (optional)
race_counts = df_pe_pos['Race_Ethnicity'].value_counts()
races_to_include = race_counts[race_counts >= 5].index.tolist()  # At least 5 patients
df_pe_pos_filtered = df_pe_pos[df_pe_pos['Race_Ethnicity'].isin(races_to_include)].copy()

print(f"Analyzing {len(df_pe_pos_filtered)} PE+ patients across {len(races_to_include)} racial/ethnic groups")

# Perform Kruskal-Wallis test
race_groups = [group['Additional_Procedures'].dropna().values 
               for name, group in df_pe_pos_filtered.groupby('Race_Ethnicity')]
h_stat, p_value = stats.kruskal(*race_groups)

print(f"\nKruskal-Wallis Test: H = {h_stat:.3f}, p = {p_value:.4f}")

# Create the figure
fig, ax = plt.subplots(figsize=(14, 8))

# Define color palette
colors = sns.color_palette("Set2", n_colors=len(races_to_include))

# Create box plot
box_parts = ax.boxplot([df_pe_pos_filtered[df_pe_pos_filtered['Race_Ethnicity'] == race]['Additional_Procedures'].values 
                         for race in races_to_include],
                        positions=range(len(races_to_include)),
                        widths=0.6,
                        patch_artist=True,
                        showfliers=False,  # We'll show outliers with swarm plot
                        medianprops=dict(color='black', linewidth=2.5),
                        boxprops=dict(facecolor='lightgray', alpha=0.7, edgecolor='black', linewidth=1.5),
                        whiskerprops=dict(color='black', linewidth=1.5),
                        capprops=dict(color='black', linewidth=1.5))

# Overlay swarm plot (individual data points)
for i, race in enumerate(races_to_include):
    race_data = df_pe_pos_filtered[df_pe_pos_filtered['Race_Ethnicity'] == race]['Additional_Procedures'].values
    
    # Add jitter to x-coordinates
    y = race_data
    x = np.random.normal(i, 0.08, size=len(y))  # Small jitter
    
    ax.scatter(x, y, alpha=0.6, s=40, color=colors[i], edgecolors='black', linewidth=0.5, zorder=3)

# Add reference line for "excessive diagnostics" threshold
ax.axhline(y=1.5, color='#e74c3c', linestyle='--', linewidth=2.5, 
           label='Excessive Diagnostics Threshold (>1 additional test)', zorder=1)

# Add text annotation for the threshold
ax.text(len(races_to_include) - 0.5, 1.6, 'Excessive Threshold', 
        fontsize=10, color='#e74c3c', fontweight='bold', ha='right')

# Customize plot
ax.set_xlabel('Race/Ethnicity', fontsize=13, fontweight='bold')
ax.set_ylabel('Number of Additional Diagnostic Procedures\n(Beyond CTPA)', fontsize=13, fontweight='bold')
ax.set_title('Diagnostic Procedure Intensity by Race/Ethnicity in PE-Positive Patients', 
             fontsize=15, fontweight='bold', pad=20)

# Set x-tick labels
ax.set_xticks(range(len(races_to_include)))
# Wrap long labels if needed
wrapped_labels = [race.replace(' ', '\n') if len(race) > 15 else race for race in races_to_include]
ax.set_xticklabels(wrapped_labels, fontsize=10, rotation=0)

# Set y-axis
ax.set_ylim(-0.2, df_pe_pos_filtered['Additional_Procedures'].max() + 0.5)
ax.set_yticks(range(0, int(df_pe_pos_filtered['Additional_Procedures'].max()) + 2))
ax.grid(axis='y', alpha=0.3, linestyle='--', zorder=0)

# Add sample sizes below x-axis
for i, race in enumerate(races_to_include):
    n = len(df_pe_pos_filtered[df_pe_pos_filtered['Race_Ethnicity'] == race])
    ax.text(i, -0.15, f'n={n}', ha='center', va='top', fontsize=9, fontweight='bold')

# Add statistical test result
ax.text(0.02, 0.98, f'Kruskal-Wallis H = {h_stat:.2f}, p = {p_value:.4f}', 
        transform=ax.transAxes, fontsize=11, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

# Add legend
ax.legend(loc='upper right', fontsize=10, frameon=True, shadow=True)

# Background
ax.set_facecolor('#f8f9fa')

plt.tight_layout()
plt.savefig('Figure2_DiagnosticIntensity_by_Race.png', dpi=300, bbox_inches='tight')
print("✓ Figure 2 saved as 'Figure2_DiagnosticIntensity_by_Race.png'")
plt.show()

# Print summary statistics for Figure 2
print("\nFigure 2 Data Summary:")
summary_stats = df_pe_pos_filtered.groupby('Race_Ethnicity')['Additional_Procedures'].agg([
    ('N', 'count'),
    ('Mean', 'mean'),
    ('SD', 'std'),
    ('Median', 'median'),
    ('Q1', lambda x: x.quantile(0.25)),
    ('Q3', lambda x: x.quantile(0.75)),
    ('Min', 'min'),
    ('Max', 'max')
]).round(2)

print(summary_stats)
```
